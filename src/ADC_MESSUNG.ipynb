# DAC Bereich durchlaufen

```python
dutV = list()
realV = list() 
diffList = list()
stepList = list()
vSet = 0.0

for step in range(0,51, 1 ):
    
   vSet = (0.1 * step);
   dp832a.setVoltage(1, vSet)
   
   vr = dmm6500.getVoltDC()
   vd = asmu.getVoltage() 
   
   if( math.isnan(vd) == True ):
       vd = asmu.getVoltage() 
        
   vd = adj(vd)  

   #diff = v - ref
   diff = (vr-vd) * 1000; # in mV


   print( "%2.2f;%2.6f;%2.6f;%3.3f" % (vSet , vr, vd ,diff))
   diffList.append( diff);
   stepList.append( vSet);
   realV.append(vr);
   dutV.append(vd);
        

```

```python

def adj(vCalc):
    p1_xr = 0.00576;
    p1_yc = 0.006516;
    p2_xr = 5.006949;
    p2_yc = 5.0113;
    
    m = ( p2_yc - p1_yc ) / (p2_xr - p1_xr);
  
    b =  p1_yc - ( m * p1_xr);

    adj = (vCalc - b ) / m;
    
    return adj
```
